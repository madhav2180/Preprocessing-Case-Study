# Outlier Removal

```python
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25) 
    Q3 = df[column].quantile(0.75) 
    IQR = Q3 - Q1  # Interquartile range
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
numeric_columns = df.select_dtypes(include=[np.number]).columns
for column in numeric_columns:
    df = remove_outliers_iqr(df, column)

print(f"Shape after outlier removal: {df.shape}")
```

```python

def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]


df_cleaned = df.copy()
for column in df.select_dtypes(include=[np.number]).columns:
    df_cleaned = remove_outliers_iqr(df_cleaned, column)


plt.figure(figsize=(16, 8))


plt.subplot(1, 2, 1)
df.select_dtypes(include=[np.number]).boxplot()
plt.title("Boxplot Before Outlier Removal", fontsize=14)
plt.xticks(rotation=45)


plt.subplot(1, 2, 2)
df_cleaned.select_dtypes(include=[np.number]).boxplot()
plt.title("Boxplot After Outlier Removal", fontsize=14)
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
